```python
df_returns = df[['Returns_pct1', 'Returns_pct2', 'Returns_pct3', 'Returns_pct4', 'Returns_pct5', 'Returns_pct6']]
df_returns.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier , StackingClassifier
```

```python
# Splitting the data into X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = train_test_split(df_returns, Y, test_size=0.30, random_state=42)
```

```python
X_train.head()
```

```python
y_train
```

```python
# Create the model with 100 trees

model = RandomForestClassifier(n_estimators=100, max_depth = 4,
                               bootstrap = True,
                               max_features = 'sqrt')
```

```python
# fitting with train data

model.fit(X_train, y_train)
```

```python
# Getting the score of the training data

model.score(X_train, y_train)
```

```python
# Getting the score of the testing data

model.score(X_test, y_test)
```

```python
# predicting the values for Xbag_train

prediction_random = model.predict(X_test)
```

```python
#crosstab

ct = pd.crosstab(y_test, prediction_random)
print(ct)
```

```python
# Classification report of the y2_predict, y2_test
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(prediction_random,y_test))
```
